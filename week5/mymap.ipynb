```python
import folium
inage =[35.617708355781346, 140.06382442208874]
port = [35.599380768884124, 140.09865995092426]
odaiba =[35.63021448643912, 139.7757180219758]
skytree =[35.710193374060424, 139.81073795092428]
shibuya =[35.65862094479192, 139.70216896441775]
m = folium.Map (location=inage and port and odaiba and skytree and shibuya,zoom_start=15)
folium. Marker(inage, popup="Inage Seaside Park, Best place to enjoy sunsets with beach", tooltip="Inage Seaside Park").add_to(m)
folium. Marker(port, popup="A nice place to watch sunsets and have picnic with friends", tooltip="Chiba Port Park").add_to(m)
folium. Marker(odaiba, popup="A great place to enjoy sunsets and city life", tooltip="Odaiba Seaside Park").add_to(m)
folium. Marker(skytree, popup="Tallest tower in Japan", tooltip="Tokyo Skytree").add_to(m)
folium. Marker(shibuya, popup="Serves you with an amazing view of the city as well as beautiful sunsets",tooltip="Shibuya Sky").add_to(m)
m
```
